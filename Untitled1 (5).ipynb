```python
from math import radians, sin, cos, sqrt, atan2

# Calculate trip distance based on coordinates

# Function to calculate Haversine distance
def haversine_distance(row):
    """
    Calculate the Haversine distance between pickup and dropoff coordinates.
    """
    # Earth radius in kilometers
    R = 6371.0
    
    # Convert latitude and longitude from degrees to radians
    lat1 = radians(row['pickup_latitude'])
    lon1 = radians(row['pickup_longitude'])
    lat2 = radians(row['dropoff_latitude'])
    lon2 = radians(row['dropoff_longitude'])
    
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# Calculate trip distance using Haversine distance
df['trip_distance_km'] = df.apply(haversine_distance, axis=1)

# Convert kilometers to miles
df['trip_distance_miles'] = df['trip_distance_km'] * 0.621371

# print(df.head())
```

```python
# Define a function to check if the pickup time is during rush hour
def is_rush_hour(pickup_time):
    # Extract the hour component from the pickup time
    # hour = pickup_time.hour
    
    # Check if the hour is within the rush hour time ranges
    if (pickup_time >= 7 and pickup_time < 9) or (pickup_time >= 16 and pickup_time < 18):
        return 1
    else:
        return 0

# Add a new column indicating whether the pickup time is during rush hour
df['is_rush_hour'] = df['pickup_hour'].apply(is_rush_hour)
```

# Linear Regression

```python
#Passing in Transformed values as Predcitors
X = transform
y = np.log(df['trip_duration']).values
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import r2_score, mean_squared_log_error , mean_squared_error, accuracy_score

# Split the data into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
model = LinearRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Predict on the testing data
y_train_pred = model.predict(X_train)
acc_train = model.score(X_train, y_train)
y_test_pred = model.predict(X_test)
acc_test = model.score(X_test, y_test)
# Get accuracy


print("Training Accuracy:", acc_train)
print("Testing Accuracy:", acc_test)

# Evaluate the model
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
print("Training Mean Squared Error:", mse_train)
print("Testing Mean Squared Error:", mse_test)

# Calculate R-squared for training and testing data
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)
print("Training R-squared:", r2_train)
print("Testing R-squared:", r2_test)

#coefficients and intercept of the model
coefficients = model.coef_
intercept = model.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)
```
